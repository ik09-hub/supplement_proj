```python
#Import Modules
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from time import sleep
import pandas as pd


```

```python
#Set Path for WebDriver
path = r'C:\Users\ishaa\chromedriver-win64\chromedriver.exe'
service = Service(executable_path=path)


#open the browser
browser = webdriver.Chrome(service=service)
```

```python
#Load webpage
browser.get('https://amazon.com')
browser.maximize_window()
```

```python
try:
    continue_button = browser.find_element(By.XPATH, "//button[.='Continue shopping']")
    continue_button.click()
    sleep(2)
except NoSuchElementException:
    print("No 'Continue shopping' button found")
```

```python
#get the input elements
input_search = browser.find_element(By.ID, "twotabsearchtextbox")
search_button = browser.find_element(By.XPATH, "(//input[@type='submit'])[1]")
```

```python
#send the input to the webpage
input_search.send_keys("protein powder")
sleep(1)
search_button.click()
```

```python
products = browser.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")
product = products[0] if products else None
print(product.get_attribute("outerHTML"))
```

```python
product_data = []

while True:
    sleep(2)  # Let page load fully
    products = browser.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")

    for product in products:

        try:
            full_title = product.find_element(By.XPATH, ".//h2[@aria-label]").get_attribute("aria-label")
            title = full_title.replace("Sponsored Ad - ", "").strip()
        except:
            title = ""


        try:    
            price_whole = product.find_element(By.CLASS_NAME, "a-price-whole").text.strip()
            price_fraction = product.find_element(By.CLASS_NAME, "a-price-fraction").text.strip()
            price = f"{price_whole}.{price_fraction}"        
        except:
            price = ""

        try:
            full_rating = product.find_element(By.CSS_SELECTOR, "span.a-declarative > a.a-popover-trigger").get_attribute("aria-label").strip()
            rating = full_rating.split(" out of")[0]
        except:
            rating = ""

        try:
            reviews = product.find_element(By.XPATH, ".//span[contains(@class, 'a-size-base') and contains(@class, 's-underline-text')]").text.strip().replace(',', '')
        except:
            reviews = ""

        product_data.append({
            "Title": title,
            "Price": price,
            "Rating": rating,
            "Reviews": reviews
        })

    try:
        next_button = browser.find_element(By.XPATH, "//a[contains(@class, 's-pagination-next')]")
        next_button.click()
    except:
        print("No more pages.")
        break
browser.quit()



```

```python
# Convert to DataFrame
protein_df = pd.DataFrame(product_data)

protein_df.head()
```
